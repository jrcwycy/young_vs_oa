```python
dpath = "/nfs/turbo/umms-indikar/shared/projects/RECODE/YvsO_results/updated_counts/"
file_list = glob.glob(f"{dpath}*")

result = []

for fpath in file_list:
    basename = os.path.basename(fpath).replace(".counts.tsv", "")
    tmp = pd.read_csv(fpath, sep='\t')
    tmp.columns = ['gene_id', 'gene_name', basename]
    tmp = tmp.set_index(['gene_id', 'gene_name'])
    result.append(tmp)
    
res = pd.concat(result, axis=1, ignore_index=False)
res = res.reset_index()

# combine bc1 - bc3 (technical replicates for the same Young D28 sample)
res['bc1_3'] = res[['bc1', 'bc2', 'bc3']].mean(axis=1).round().astype(int)
res.drop(columns=['bc1', 'bc2', 'bc3'], inplace=True)

print(f"{res.shape=}")
res.head()
```

# Make gene expression plots

```python
dpath = "/nfs/turbo/umms-indikar/shared/projects/RECODE/YvsO_results/updated_counts/"
file_list = glob.glob(f"{dpath}*")

p = 1000 # kilobase pair resolution on gene length

result = []

for fpath in file_list:
    basename = os.path.basename(fpath).replace(".counts.tsv", "")
    tmp = pd.read_csv(fpath, sep='\t')
    tmp.columns = ['gene_id', 'gene_name', 'read_count']
    
    tmp.drop('gene_name', axis=1, inplace=True)
    
    tmp = pd.merge(tmp, gdf, 
                   how='left', 
                   left_on='gene_id',
                   right_on='gene_id')
    
    # filter missing mappings
    tmp = tmp[tmp['gene_name'].notna()].reset_index(drop=True)

    # normalize expression (TPM)    
    tmp['RPK'] = tmp['read_count'] / (tmp['Length'] / p)
    scaling_factor = (tmp['RPK'].sum())/1e6
    tmp['TPM'] = tmp['RPK']/scaling_factor

    tmp['file'] = basename

    tmp = tmp[['gene_id', 'gene_name', 'gene_biotype', 'Length', 'file', 'read_count', 'TPM']]
    
    result.append(tmp)

    
result = pd.concat(result)

# Remove 0 read counts
result = result[result['read_count'] > 0]

# compute other metrics for plotting
result['log2(TPM+1)'] = np.log2(result['TPM'] + 1)
result['log(TPM+1)'] = np.log1p(result['TPM'])

print(f"{result.shape=}")
result.head()
```

```python
res = pd.merge(result, samples,
               how='left',
               left_on='file',
               right_on='Barcode')

mapper1 = {'Young_D28' : 'Young', 'OA_D28' : 'OA', 'Young_D14' : 'Young', 'OA_D14' : 'OA'}
mapper2 = {'Young_D28' : 'D28', 'OA_D28' : 'D28', 'Young_D14' : 'D14', 'OA_D14' : 'D14'}

# add condition and timepoint columns for plotting
res['Condition'] = res['Group'].map(mapper1)
res['Time'] = res['Group'].map(mapper2)

print(f"{res.shape=}")
res.head()
```

```python
df = res.copy()

values_to_plot = 'log2(TPM+1)' # Set to whichever normalization method you want to plot (column in df)

# filter for protein-coding genes
df = df[df['gene_biotype'] == 'protein_coding']
print(f"{df.shape=}")

df = df.pivot(index='gene_name', columns='file', values=values_to_plot)
df = df.fillna(0)

# Take average of bc1-bc3 since technical replicate
df['bc1_3'] = df[['bc1', 'bc2', 'bc3']].mean(axis=1)
df.drop(columns=['bc1', 'bc2', 'bc3'], inplace=True)

# Add sample info for plotting
df = df.T

day = ['D28'] * 2 + ['D14'] + ['D28'] + ['D14'] + ['D28'] + ['D14'] * 4 + ['D28'] * 3
group = ['OA'] * 2 + ['Young'] * 2 + ['OA'] * 2 + ['Young'] * 2 + ['OA'] * 2 + ['Young'] * 3

sample_info = pd.DataFrame({
    'Sample': df.index,
    'Group': group,
    'Day': day
})

df = df.merge(sample_info, left_index=True, right_on='Sample')

print(f"{df.shape=}")
df.head()
```

```python
def plot_genes(genes, max_num_genes, data):
    """
    Plots specific genes of interest in a specified subplot orientation, with 1 row, for YvsO data 
    
    Param: genes: A list of genes to plot
    Param: max_num_genes: Max number of genes in gene list group; Sets number of columns in 1 row subplot
    Param: data: A dataframe with genes as columns, rows as samples/files (e.g. barcode), and values as normalized gene expression
    
    NOTE: Set figsize prior to running function
    """

    colors = ['#3E7274', '#FF8080']
    err_kws = {'color' : 'k', 'linewidth' : 1}
    
    for i in range(max_num_genes):
        plt.subplot(1, max_num_genes, i + 1)
        
        if i < len(genes): # Plot only if the gene exists
            gene = genes[i]
            if gene in data.columns:
                sns.barplot(data=data,
                            x='Day',
                            y=gene,
                            edgecolor='k',
                            linewidth=1.5,
                            errorbar='sd',
                            capsize=0.2,
                            hue='Group',
                            palette=colors,
                            hue_order=['OA', 'Young'],
                            order=['D14', 'D28'],
                            err_kws=err_kws,
                            legend=False)
                plt.title(gene, fontsize=12, fontweight='bold')
                
            else:
                plt.title(gene, fontsize=12, color='r')
                #plt.axis('off')
                
        else:
            plt.axis('off') # Hide axis if there is no gene
            
        plt.xlabel('')
        plt.ylabel('log2(TPM+1)', fontsize=12)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=10)
        plt.gca().spines['left'].set_linewidth(1.5)
        plt.gca().spines['bottom'].set_linewidth(1.5)
        plt.gca().tick_params(axis='both', width=1.5)
        
    
    sns.despine()
    plt.tight_layout()
    plt.show()
```

```python
chondro_genes = ["SOX9", "SOX5", "SOX6", "COL2A1", "ACAN", "COMP", "SNORC", "COL11A1"]

plt.rcParams['figure.dpi'] = 300
plt.figure(figsize=(18, 3))

plot_genes(chondro_genes, 8, df)
```

```python
hypertrophic_genes = ["COL1A1", "COL10A1", "RUNX2", "MMP13", "MEF2C", "SMAD1", "SMAD5", "SMAD9"]

plt.rcParams['figure.dpi'] = 300
plt.figure(figsize=(18, 3))

plot_genes(hypertrophic_genes, 8, df)
```

# Dot plot

This code plots a dotplot of z-scored average expression for a defined set of genes across conditions. 
* z-score = 0: at the mean
* z-score > 0: above the mean
* z-score < 0: below the mean

Dot size is 1 / Coefficient of Variation (CV) to capture relative variability across biological replicates per condition. 
* Low values = more variable 
* High values = more stable

Dots that are missing from the plot indicate that gene is not expressed in at least 2 biological replicates for a condition.

```python
plot_df = res.copy()

# Choose genes for plotting here
genes_to_plot = [
    # chondrogenic
    'SOX9', 'COL2A1', 'ACAN', 'COMP', 'PRG4', 'SNORC', 'COL9A1', 'COL11A1', 'MATN3', 'GDF5',
    # hypertrophic
    'COL1A1', 'COL10A1', 'RUNX2', 'IHH', 'MMP13', 'SMAD1', 'SMAD5', 'SMAD9',
]

# filter for genes of interest
plot_df = plot_df[plot_df['gene_name'].isin(genes_to_plot)]

# collapse bc1-bc3 (technical replicates for Young D28)
plot_df['BioRep'] = plot_df['Barcode']
plot_df.loc[plot_df['Barcode'].isin(['bc1', 'bc2', 'bc3']), 'BioRep'] = 'rep1'

collapsed = (
    plot_df.groupby(['gene_name', 'Group', 'BioRep'], as_index=False)
    .agg({'TPM': 'mean'})
)

# get means and standard deviations of TPM across replicates
group_df = (
    collapsed.groupby(['gene_name', 'Group'])
    .agg(mean_TPM=('TPM', 'mean'),
         sd_TPM=('TPM', 'std'))
    .reset_index()
)

# compute coefficient of variance
group_df['cv_TPM'] = group_df['sd_TPM'] / (group_df['mean_TPM'] + 1e-6) # avoid div by 0
group_df['stability'] = 1 / (group_df['cv_TPM'] + 1e-6)

# Pivot for plotting matrices
avg_expr = group_df.pivot(index='gene_name', columns='Group', values='mean_TPM')
stability = group_df.pivot(index='gene_name', columns='Group', values='stability')
stability = stability.fillna(0)

# Z-score mean TPM per gene (across groups)
avg_expr_z = avg_expr.apply(lambda x: scipy.stats.zscore(x, nan_policy='omit'), axis=1)

# Put results back into DataFrame with same shape
avg_expr_z = pd.DataFrame(avg_expr_z.tolist(),
                          index=avg_expr.index,
                          columns=avg_expr.columns)


# Define order of groups on the y-axis here (first group is on the bottom)
groups = ['OA_D14', 'OA_D28', 'Young_D14', 'Young_D28']


cmap = 'RdBu'
size_scale = 20 
vmin = -2
vmax = 2

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 10, 3
fig, ax = plt.subplots()

for i, group in enumerate(groups):    # y-axis = groups  
    for j, gene in enumerate(genes_to_plot):  # x-axis = genes
        size = stability.loc[gene, group] * size_scale
        color = avg_expr_z.loc[gene, group]

        ax.scatter(
            x=j, y=i,
            s=size,
            c=[color],
            # ec='k', # uncomment these 2 lines to add a black outline to each dot
            # lw=1,
            cmap=cmap,
            vmin=vmin, vmax=vmax
        )

# Set ticks
ax.set_xticks(range(len(genes_to_plot)))
ax.set_xticklabels(genes_to_plot, rotation=90)

group_labels = ['OA (D14)', 'OA (D28)', 'Young (D14)', 'Young (D28)']
ax.set_yticks(range(len(groups)))
ax.set_yticklabels(group_labels)

# Axis labels
ax.set_ylabel('')
ax.set_xlabel('')

ax.margins(y=0.2) # adds margin to y-axis 


# ### Legend for dot size with numeric values
# sizes = [1, 5, 8, 12] # example values based on min max of stabilities
# markers = [plt.scatter([], [], s=s*size_scale, color='gray') for s in sizes]
# labels = [f"{s}" for s in sizes]

# ax.legend(markers, labels, title="1/Coeff. of Variation", scatterpoints=1,
#           frameon=False, bbox_to_anchor=(1.01, 1), loc='upper left')


### Legend for dot size with text labels
sizes = [1, 10] # example values based on min max of stabilities. May need to adjust based on actual stability values.
markers = [plt.scatter([], [], s=s*size_scale, color='gray') for s in sizes]
labels = ['less stable', 'stable']

ax.legend(markers, labels, title="1/Coeff. of Variation", scatterpoints=1,
          frameon=False, bbox_to_anchor=(1.01, 1), loc='upper left')


plt.tight_layout()
plt.show()
```

```python
plt.rcParams['figure.dpi'] = 300
fig, ax = plt.subplots(figsize=(1, 2.5))

cbar = fig.colorbar(
    plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax)),
    cax=ax, orientation='vertical'
)

ax.set_position([0.2, 0.1, 0.1, 0.6]) # [left, bottom, width, height] in figure coordinates
cbar.set_label("z-scored avg TPM", fontsize=11)
cbar.ax.tick_params(labelsize=9)

plt.show()
```

# Load DESeq2 Results

```python
def sig_deseq2_results(path, comparison, padj_thresh=0.05, lfc_thresh=1):
    
    df = pd.read_csv(path)
    cols = df.columns.to_list()
    cols[0] = 'gene_id'
    df.columns = cols
    
    df = pd.merge(df, gdf,
                  how='left',
                  left_on='gene_id',
                  right_on='gene_id',
                 )
    
    #df = df[df['gene_biotype'] == 'protein_coding'] # Filter for protein-coding genes
    df = df[df['gene_name'].notna()].reset_index(drop=True) # Drop unnamed genes
    
    df = df[df['padj'] < padj_thresh] # Filter for adjusted p-value < 0.05
    df = df[abs(df['log2FoldChange']) > lfc_thresh] # Filter for Log2FC > 1
    
    print(f"Number of significant DE genes for {comparison}: {df.shape[0]}") # Print number of DE genes
    
    return df


def load_pathway(fpath):
    """
    Loads an Enrichr-like database file into a boolean DataFrame.

    Args:
        fpath (str): Path to the Enrichr-like database file.

    Returns:
        pandas.DataFrame: A boolean DataFrame where:
            - Index: Genes
            - Columns: Pathways
            - Values: True if the gene is in the pathway, False otherwise.
    """

    result = []
    with open(fpath) as f:
        for line in f:
            split_line = [x for x in line.strip().split('\t') if x]  # Remove empty strings directly

            row = {'label': split_line[0]}
            for gene in split_line[1:]:
                row[gene] = 1

            result.append(row)

    df = pd.DataFrame(result)
    df = df.fillna(0.0).set_index('label').astype(bool).T  # Chained operations for clarity

    return df


def getGenes(pdf, cellType, ui_upper=None):
    genes = pdf[pdf['cell type'] == cellType]
    
    if not ui_upper is None:
        genes = genes[genes['ubiquitousness index'] < ui_upper]
        
    return genes['official gene symbol'].to_list()


def get_chon_genes(df, gene_list, comparison):
    """
    Filters a df for specific genes and splits df into up and down regulated genes, each in a new df
    
    df: Dataframe with DESeq2 results
    gene_list: List of genes you want to filter for
    comparison: Name for printing the # of DE genes
    
    Returns: The original df filtered for genes in gene_list, the upregulated genes, and the downregulated genes
    """
    
    tmp = df[df['gene_name'].isin(gene_list)]
    print(f"No. DE Chondrocyte genes for {comparison}: {tmp.shape[0]}")
    
    tmp = tmp.reindex(tmp['log2FoldChange'].abs().sort_values(ascending=False).index)

    tmp_up = tmp[tmp['log2FoldChange'] > 0]
    tmp_down = tmp[tmp['log2FoldChange'] < 0]

    print(f"UP genes for {comparison}: {tmp_up.shape[0]}")
    print(f"DOWN genes for {comparison}: {tmp_down.shape[0]} \n")
    
    return tmp, tmp_up, tmp_down
```

```python
def plot_pathway_enrichment(df, color, ax):
    """
    Plots pathway enrichment results from a DataFrame.

    Args:
        df: DataFrame containing 'path_name' and 'p_val' columns.
    """
    
    # Sort and calculate -log10(p_val)
    df_sorted = df.sort_values('adj_p_val')
    df_sorted['neg_log10_pval'] = -np.log10(df_sorted['adj_p_val'])
    df_sorted['pname'] = df_sorted['path_name'].apply(lambda x: x.split("(")[0])#.capitalize())
    #df_sorted['pname'] = df_sorted['Term'].apply(lambda x: x.split(" WP")[0])#.capitalize()) # For WikiPathways
    
    wrapped_labels = [fill(label, 50) for label in df_sorted['pname']]

    
    #Create the plot
    ax.barh(
        wrapped_labels, 
        df_sorted['neg_log10_pval'], 
        color=color,
        height=0.5,
        ec='k',
    )
    # sns.barplot(
    #     wrapped_labels,
    #     df_sorted['neg_log10_pval'], 
    #     color=color,
    #     height=0.5,
    #     edgecolor='k',
    # )

    # Set labels and title
    ax.set_xlabel('-log10(adjusted p-value)', fontsize=16)
    ax.set_yticks(np.arange(len(wrapped_labels)))
    ax.set_yticklabels(wrapped_labels, fontsize=16)
    
    #plt.yticks(fontsize=14)

    # Invert y-axis and set x-axis ticks/limits
    ax.invert_yaxis()
    sns.despine(ax=ax)

    # Show the plot
    plt.tight_layout()
```
